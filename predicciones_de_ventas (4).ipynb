Este grÃ¡fico nos muestra que no se necesita mucha visibilidad para generar buenas ventas. Los productos con mucha visibilidad no cumplen los estandares de venta. 

```python
#Acumulacion de ventas

cumulototaldeventas = df.groupby('Item_Type')['Item_Outlet_Sales'].sum() / 1000
cumulototaldeventas.sort_values(inplace = True)

plt.figure(figsize = (22, 9));
plt.bar(cumulototaldeventas.index, cumulototaldeventas, width = 0.9, color= 'brown');
plt.xticks(color= 'brown', rotation = 40);
plt.xlabel('Tipos de productos', color= 'brown', fontsize = 16);
plt.ylabel('Ventas por miles',color= 'green', fontsize = 16);
plt.title('Ventas por tipo de producto', fontsize = 20);
```

Este gr[afico demuestra que lo menos vendido en la tienda son la comida de mar y los desayunos. Mientras que las fruta, snacks y todo lo que tiene que ver con cosas de casa son las mas apetecidas.
